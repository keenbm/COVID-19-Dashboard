```python
state_daily=pd.read_csv("state_wise_daily.csv",parse_dates=['Date'])
state_daily.drop(['TT'],axis=1,inplace=True)
state_daily.head()
```

```python
## Converting into 3 different DF for doing running sum
Confirmed_df=state_daily.query("Status=='Confirmed'")
Recovered_df=state_daily.query("Status=='Recovered'")
Deceased_df=state_daily.query("Status=='Deceased'")

## Shorting by Date to maintain sequence
Confirmed_df.sort_values(by=['Date'],inplace=True)
Recovered_df.sort_values(by=['Date'],inplace=True)
Deceased_df.sort_values(by=['Date'],inplace=True)

## Applyinh Running run in all three data Frame
Confirmed_df.loc[:,"AN":]=Confirmed_df.loc[:,"AN":].cumsum(axis = 0)
Recovered_df.loc[:,"AN":]=Recovered_df.loc[:,"AN":].cumsum(axis = 0)
Deceased_df.loc[:,"AN":]=Deceased_df.loc[:,"AN":].cumsum(axis = 0)

## Stacking all three dataframe
comb_new=pd.concat([Confirmed_df,Recovered_df, Deceased_df], ignore_index=True)
comb_new
```

```python
state_daily_new=comb_new.melt(id_vars=["Date","Status"], 
        var_name='State_Code',
        value_name="Count")
state_daily_new
```

```python
temp=state_daily_new.pivot_table(values = 'Count', 
                      index = ['Date','State_Code'], 
                      columns = 'Status', 
                      aggfunc = 'sum')
temp.to_excel('temp.xlsx')
```
